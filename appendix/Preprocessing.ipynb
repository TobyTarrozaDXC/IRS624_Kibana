```python
import pandas as pd 
import numpy as np 

movies = pd.read_csv('IMDb movies.csv')
```

```python
movies
```

```python
movies_trimmed = movies[['imdb_title_id', 'original_title', 'year', 'genre', 'duration', 'country', 
                         'language', 'director', 'writer', 'production_company', 'actors', 'description', 'avg_vote',]
                       ].dropna(subset=['writer','year', 'duration', 'avg_vote'])
```

```python
movies_trimmed.info()
```

```python
print(movies_trimmed["year"].min(), movies_trimmed["year"].max())
print(movies_trimmed["duration"].min(), movies_trimmed["duration"].max())
print(movies_trimmed["avg_vote"].min(), movies_trimmed["avg_vote"].max())
genre_list = movies_trimmed["genre"].unique()
genre_unique = [ x for x in genre_list if " " not in x ]
print(genre_unique)
```

```python
country_list = movies_trimmed["country"].unique()
country_unique = [ x for x in country_list if " " not in x ]
print(country_unique)
```

```python
language_list = movies_trimmed["language"].unique()
language_unique = [ x for x in language_list if " " not in x ]
print(language_unique)
```

```python
movies_trimmed[:10000].to_csv('movies_trimmed1.csv', index=False)
movies_trimmed[10000:20000].to_csv('movies_trimmed2.csv', index=False)
movies_trimmed[20000:30000].to_csv('movies_trimmed3.csv', index=False)
movies_trimmed[30000:40000].to_csv('movies_trimmed4.csv', index=False)
movies_trimmed[40000:50000].to_csv('movies_trimmed5.csv', index=False)
movies_trimmed[60000:70000].to_csv('movies_trimmed6.csv', index=False)
movies_trimmed[70000:].to_csv('movies_trimmed7.csv', index=False)
```
